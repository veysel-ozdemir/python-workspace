```python
import pandas as pd

df = pd.read_csv('AirPassengers.csv')

df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m')
df.head()
```

```python
df.index=df['date']
df.head()
```

```python
del df['date']
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.lineplot(df)
plt.ylabel('Passenger Count')
plt.xlabel('Years')
plt.title('Passenger count of years')
plt.show()
```

```python
ortalama7 = df.rolling(7).mean()
std_7 = df.rolling(7).std()

plt.plot(df, color='blue', label='Original Data')
plt.plot(ortalama7, color='red', label='Average (7 Month)')
plt.plot(std_7, color='green', label='Standard Deviation (7 Month)')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

adftest = adfuller(df, autolag='AIC')

adftest
```
