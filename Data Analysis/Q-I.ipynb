```python
import pandas as pd
```

```python
NUM = 11
```

```python
students = pd.read_csv('student_dataset.csv')
```

```python
# İlk 5 değer
students.head(5)
```

```python
# Son 5 değer
students.tail(5)
```

```python
# Gender değeri Female ve Lunch değeri Free olan öğrencilerin sayısı
total_female_free = students.loc[(students['Gender'] == 'Female') & (students['Lunch'] == 'Free')]['Student_ID'].count()
total_female_free
```

```python
# Gender değeri Female olan öğrencilerin;
# Student_ID ve Math_Score değerlerinden oluşan yeni bir DataFrame
female_id_mathScore = students[['Student_ID', 'Math_Score']][students['Gender'] == 'Female']
female_id_mathScore.head(5)
```

```python
# Writing_Score değeri NUM değerinden (NUM = 11) küçük olan öğrencilerin;
# Student_ID, Gender ve Writing_Score değerlerinden oluşan yeni bir DataFrame
low_writingScore = students[['Student_ID', 'Gender', 'Writing_Score']][students['Writing_Score'] < NUM]
low_writingScore
```

```python
# students DataFrame'in NUM nolu (NUM = 11) satırın Math_Score sütununun ilk değeri
students.loc[NUM, 'Math_Score']
```

```python
# students DataFrame'in NUM nolu (NUM = 11) satırın Math_Score değerini NUM değeri ile değiştirme
students.loc[NUM, 'Math_Score'] = NUM
```

```python
# students DataFrame'in NUM nolu (NUM = 11) satırın Math_Score sütununun son değeri
students.loc[NUM, 'Math_Score']
```

```python
# Belirtilen kurala göre, students DataFrame'ine
# Average isimli yeni bir sütun oluşturma
students['Average'] = students['Math_Score'] * 0.4 + students['Reading_Score'] * 0.3 + students['Writing_Score'] * 0.3
students.head(5)
```

```python
# Average sütunundaki değerlerinin ortalaması
students['Average'].mean()
```

```python
# Average değeri tüm öğrencilerin Average değerinin ortalamasından küçük öğrencilerin;
# bütün bilgilerinden oluşan yeni bir DataFrame oluşturma
low_average_students = students.loc[students['Average'] < students['Average'].mean()]
low_average_students.head(5)
```

```python
# students DataFrame'ini Gender'a göre gruplandırarak,
# Test_Preparation_Course sütunundaki değerlerin sayısını bulma
gender_course_count = students.groupby(['Gender'])[['Test_Preparation_Course']].count()
gender_course_count
```
