```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = [[1, 5], [5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35]]
x = np.array(x)

y = [4, 5, 20, 14, 32, 22, 38, 43]
y = np.array(y)

fx = LinearRegression().fit(x, y)
print(f'Intercept (b0): {fx.intercept_}')
print(f'Coefficents (b1, b2): {fx.coef_}')
print('b1: ', fx.coef_[0])
print('b2: ', fx.coef_[1])
print(f'R2 score: {fx.score(x, y)}')

# f(x) = 4.75 + 0.49x1 + 0.18x2

fx_pred = fx.predict(x)
print(f'\nPredicted: {fx_pred}')

y_pred = fx.intercept_ + np.sum(fx.coef_ * x, axis=1) # axis=1: islemleri sütunlara göre yap
print(f'\nPredicted: {y_pred}')
```

```python
# example I
value = np.array([[6, 7], [30, 5], [85, 45]]) # no need to reshape
print(value)

value_pred = fx.predict(value)
print(f'predicted: {value_pred}\n')

# example II
arr = np.arange(10).reshape(-1, 2) # belirlenen sütun sayisina göre satiri otomatik ayarla
print(arr)

new_pred = fx.predict(arr)
print(f'predicted: {new_pred}\n')

# example III
b0 = fx.intercept_
b1 = fx.coef_[0]
b2 = fx.coef_[1]

x1 = 0
x2 = 1
# x = arr[0][:] # get the items of 1st row

model = b0 + b1*x1 + b2*x2
# model = b0 + np.sum([b1, b2] * x)
print(model)
```
