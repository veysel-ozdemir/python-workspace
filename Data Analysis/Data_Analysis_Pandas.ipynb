```python
notes = pd.read_csv("ogrenci.csv")
notes.head()
```

```python
notes["ort"] = notes["vize"]*0.4 + notes["final"]*0.6
notes.head()
```

```python
notes.describe()
```

```python
# Solution 1
passed_students = notes[["ad", "ort"]][notes["ort"] >= notes["ort"].mean()] # same as notes[notes["ort"] >= notes["ort"].mean()][["ad", "ort"]]
passed_students.head()
```

```python
# Solution 2
passed_std = notes.loc[notes["ort"] >= notes["ort"].mean(), ["ad", "ort"]]
passed_std.head()
```
