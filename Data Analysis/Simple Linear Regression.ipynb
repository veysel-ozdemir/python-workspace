```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# in order to perform multiplication on matrices we have to convert x to a column vector
x = np.array([5, 15, 25, 35, 45, 55]).reshape(-1, 1) # -1: satir sayisini otomatik olarak ayarlayan deger
y = np.array([5, 20, 14, 32, 22, 38])

fx = LinearRegression().fit(x, y)
print("B0: ", fx.intercept_)
print("B1: ", fx.coef_) # dogrunun egimi
print("R2 score: ", fx.score(x, y))

# f(x) = 5.63 + 0.54x

y_pred = fx.predict(x)
print(f"y_pred = {y_pred}\n")

# visualizing graphically
plt.scatter(x, y, color='red', marker='o', s=30) # nokta grafigi olusturmak icin kullanilir
plt.plot(x, y_pred, color='orange') # dogruyu ciz
plt.show() # göster

# examples
f30 = fx.intercept_ + fx.coef_ * 30
print(f"\nf(30) = {f30}")

otherway = fx.predict(np.array([30]).reshape(-1, 1))
print(f"f(30) = {otherway}")

vals = [10, 20, 30, 40, 50]
print(f"values = {vals}")
pred_vals = fx.predict(np.array(vals).reshape(-1, 1))
print(f"predicted values = {pred_vals}")
```

```python
x = np.array([24, 30, 28, 45, 43, 19, 55, 36]).reshape(-1, 1)
y = np.array([12000, 23000, 18000, 30000, 34000, 17000, 60000, 21000])

fx = LinearRegression().fit(x, y)
print("B0: ", fx.intercept_)
print("B1: ", fx.coef_)
print("R2 score: ", fx.score(x, y))

# f(x) = -12672.24 + 1229.92x

y_pred = fx.predict(x)
print(f"y_pred = {y_pred}\n")

fx33 = 33
fy33 = fx.intercept_ + fx.coef_ * 33 # predicted
print(f"f(33) = {fy33}\n")

yeni_personel = [34, 41, 18, 54]
yeni_maas = fx.predict(np.array(yeni_personel).reshape(-1, 1))
print(f"yeni_personel = {yeni_personel}")
print(f"yeni_maas = {yeni_maas}\n")

plt.scatter(x, y, color='red', marker='o', s=30)
plt.scatter(fx33, fy33, color='green', marker='+', s=30) # belli bir noktayi grafikte göstermek
plt.scatter(yeni_personel, yeni_maas, color='blue', marker='o', s=30) # eklenen yeni veriler
plt.plot(x, y_pred, color='orange')
plt.show() # göster
```
