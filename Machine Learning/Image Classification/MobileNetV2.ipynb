### ğŸš€ **Classifying Images with MobileNetV2**

### ğŸ“¦ Importing Pre-trained MobileNetV2

```python
from keras.applications.mobilenet_v2 import MobileNetV2

model = MobileNetV2(weights="imagenet")
model.summary()
```

### ğŸ“· Loading the image

```python
import numpy as np
from imageio import imread

data = np.empty((1, 224, 224, 3)) # 1 batch, 224x224x3
data[0] = imread("images/cat.jpg")
```

```python
print(data[0]) # The image is not normalized
print(f'min: {data[0].min()}, max: {data[0].max()}')
```

### ğŸ”„ Normalizing the image

```python
from keras.applications.mobilenet_v2 import preprocess_input

data = preprocess_input(data)
```

```python
print(data[0]) # The image is normalized
print(f'min: {data[0].min()}, max: {data[0].max()}')
```

### âœ… Classifying the image

```python
predictions = model.predict(data)
```

```python
predictions.shape # 1 batch, 1000 classes
```

```python
np.argmax(predictions, axis=1) # Index of the maximum value
```

```python
predictions[0][281] # Probability of the 281th class
```

### âš™ï¸ Decoding the predictions

```python
from keras.applications.mobilenet_v2 import decode_predictions

for name, desc, score in decode_predictions(predictions, top=5)[0]:
  print('- {} ({:.2f}%)'.format(desc, 100 * score))
```

### ğŸ”š Conclusion
According to the prediction results of the model, the provided ***cat.jpg*** image identifies a **tabby cat** with a percentage of **67.83%**.
